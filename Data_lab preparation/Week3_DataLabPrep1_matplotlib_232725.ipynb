```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python

plt.figure(figsize=(8, 5))
plt.hist(df['Survived'], bins=[-0.5, 0.5, 1.5], rwidth=0.8, color='skyblue', edgecolor='black')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.title('Survival Count (0 = Not Survived, 1 = Survived)')
plt.xlabel('Survival')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for Age Distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=30, color='lightcoral', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for Passenger Class
plt.figure(figsize=(8, 5))
plt.hist(df['Pclass'], bins=[0.5, 1.5, 2.5, 3.5], rwidth=0.8, color='mediumseagreen', edgecolor='black')
plt.xticks([1, 2, 3], ['1st Class', '2nd Class', '3rd Class'])
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for Gender
plt.figure(figsize=(8, 5))
plt.hist(df['Sex'], bins=[-0.5, 0.5, 1.5], rwidth=0.8, color='gold', edgecolor='black')
plt.xticks([0, 1], ['Male', 'Female'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot for Survival by Passenger Class
plt.figure(figsize=(8, 5))
survival_by_class = df.groupby(['Pclass', 'Survived']).size().unstack()
survival_by_class.plot(kind='bar', stacked=True, color=['salmon', 'lightblue'], edgecolor='black')
plt.title('Survival by Passenger Class (0 = Not Survived, 1 = Survived)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```
