```python
from sklearn.preprocessing import StandardScaler, RobustScaler 
import numpy as np
```

```python
import pandas as pd

# Load the unified dataset created in Week 3
file_path = "football_data.csv"
df = pd.read_csv("football_data.csv")
df.head()
```

```python
df[df.select_dtypes(np.number).columns] =  StandardScaler().fit_transform(df[df.select_dtypes(np.number).columns])
df[["Height","Weight"]] 
```

```python

df[df.select_dtypes(np.number).columns] =  RobustScaler().fit_transform(df[df.select_dtypes(np.number).columns])
df[["Height","Weight"]] 
```
