```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming fb is your DataFrame containing football player data

# Chart 1: Distribution of Player Ages
plt.figure(figsize=(10, 6))
sns.histplot(fb['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.savefig('Week3DataPrep1_232725_chart1.jpeg')
plt.show()

# Chart 2: Scatter Plot of Market Values vs. Ages
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Market value', data=fb, color='orange', alpha=0.5)
plt.title('Scatter Plot of Market Values vs. Ages')
plt.xlabel('Age')
plt.ylabel('Market Value')
plt.savefig('Week3DataPrep1_232725_chart2.jpeg')
plt.show()

# Chart 3: Box Plot of Player Heights (or another numerical column)
if 'Height' in fb.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=fb['Height'], color='lightblue')
    plt.title('Box Plot of Player Heights')
    plt.xlabel('Height')
    plt.savefig('Week3DataPrep1_232725_chart3.jpeg')
    plt.show()
else:
    print("Height column not found in the dataset.")

# Chart 4: Correlation Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = fb[['Age', 'Market value']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('Week3DataPrep1_232725_chart4.jpeg')
plt.show()

# Chart 5: Distribution of Player Nationalities
plt.figure(figsize=(12, 6))
nationality_counts = fb['Birth country'].value_counts()
sns.barplot(x=nationality_counts.index, y=nationality_counts.values, color='green')
plt.title('Distribution of Player Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.xticks(rotation=45, ha='right')
plt.savefig('Week3DataPrep1_232725_chart5.jpeg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming fb is your DataFrame containing football player data

# Chart 1: Distribution of Player Ages
plt.figure(figsize=(10, 6))
sns.histplot(fb['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.savefig('Week3DataPrep1_232725_chart1.jpeg')
plt.show()

# Chart 2: Scatter Plot of Market Values vs. Ages
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Market value', data=fb, color='orange', alpha=0.5)
plt.title('Scatter Plot of Market Values vs. Ages')
plt.xlabel('Age')
plt.ylabel('Market Value')
plt.savefig('Week3DataPrep1_232725_chart2.jpeg')
plt.show()

# Chart 3: Box Plot of Player Heights (or another numerical column)
if 'Height' in fb.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=fb['Height'], color='lightblue')
    plt.title('Box Plot of Player Heights')
    plt.xlabel('Height')
    plt.savefig('Week3DataPrep1_232725_chart3.jpeg')
    plt.show()
else:
    print("Height column not found in the dataset.")

# Chart 4: Bar Chart of Correlation Coefficients
plt.figure(figsize=(8, 6))
selected_attributes = ['Age', 'Market value']
correlation_matrix = fb[selected_attributes].corr()

plt.bar(correlation_matrix.columns, correlation_matrix.iloc[0], color='orange', alpha=0.7, label=correlation_matrix.index[0])
plt.bar(correlation_matrix.columns, correlation_matrix.iloc[1], color='blue', alpha=0.7, label=correlation_matrix.index[1])

plt.title('Correlation Coefficients between Age and Market Value')
plt.ylabel('Correlation Coefficient')
plt.legend()
plt.savefig('Week3DataPrep1_232725_chart4.jpeg')
plt.show()

# Chart 5: Distribution of Top 50 Player Nationalities
top_50_countries = fb['Birth country'].value_counts().nlargest(50)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_50_countries.index, y=top_50_countries.values, color='green')
plt.title('Distribution of Top 50 Player Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.xticks(rotation=45, ha='right')
plt.savefig('Week3DataPrep1_232725_chart5.jpeg')
plt.show()
```
