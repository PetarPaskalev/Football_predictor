Write a Python function that creates 2 additional columns "Start Year" and "End Year" for the "Year of Birth" column. (Consider 1900 as the lowest year, and 2023 as the highest year for first and last data point)

```python
def add2_columns(df):
    start_call = []
    end_call = []
    for i,item in enumerate(df["Year of Birth"]):
        if i == 0: 
            start_call.append(1900) 
            end_call.append(int(item[:4]))
        elif i == len(df["Year of Birth"])-1:
            end_call.append(2023) 
            start_call.append(int(item[:4]))
        else: 
            start_call.append(int(item[:4]))
            end_call.append(int(item[-4:]))
    df["Start Year"]= start_call
    df["End Year"]= end_call
    return df 
add2_columns(df)
        
```

Write a Python function that creates 2 additional columns "Age max" and "Age min" for the "Year of Birth" column. (i.e, for 1992, the Age max would be 32 and age min would be 1)

```python
def age_max_min(df):
    df= add2_columns(df)
    df["Age max"] = 2024-df["Start Year"]
    df["Age min"] = 2024-df["End Year"]
    return df 
age_max_min(df)
    
```

Write a Python function that creates an additional column "Vaccine registration date" based on "Vaccine registration" column. This has to be in date format.

```python
from datetime import datetime
def vaccine_reg_date(df):
    date_call=[]
    for item in df["Vaccine registration"]:
        date_string = item[5:]
        if date_string[0] == 'e':
            date_string = '01 June 2021'
        if date_string[0] == 'm':
            date_string = '15 June 2021'
        date_format = '%d %B %Y'
        parsed_date = datetime.strptime(date_string,date_format)
        date_call.append(parsed_date)
    df["Vaccine registration date"]= date_call
    return df 
vaccine_reg_date(df)



```

Which sex had a higher chance of surviving the disaster?

```python
female_survivors = df[df['Sex'] == 'female']['Survived'].sum()/df[df['Sex'] == 'female'].shape[0] 
male_survivors = df[df['Sex'] == 'male']['Survived'].sum()/df[df["Sex"]=="male"].shape[0]
if female_survivors > male_survivors:
    print('Female')
else:
    print('Male')
```

- Group the data into age groups (young, adult, old) and find which age group had the highest chance of survival?
    - ages less than or equal to 18 are classified as young.
    - ages greater than 18 but lesser than or equal to 60 are classified as adult.
    - ages greater than 60 are classified as old.


```python
young = df[(df['Age'] <= 18)]['Survived'].sum()/df[(df['Age'] <= 18) ].shape[0] 
adult = df[(df['Age'] > 18) & (df['Age'] <= 60)]['Survived'].sum()/df[(df['Age'] > 18) & (df['Age'] <= 60)].shape[0] 
old = df[df['Age'] > 60]['Survived'].sum()/df[df['Age'] > 60].shape[0] 

most_survivors = max([young, adult, old])

if young == most_survivors:
    print("Young")
if adult == most_survivors:
    print("Adult")
if old == most_survivors:
    print("Old")
```

# 4. Integration challenge


 Use the age column in the Titanic dataset, to map with age data from RIMV dataset that you created today. Remember we have "Age min" and "Age max" columns, so ideally you have to find out which range that each of the Titanic passenger belongs to.

```python
vaccine_data = pd.read_csv('vaccine.csv')
titanic_data = pd.read_csv('titanic.csv')
vaccine_data = age_max_min(vaccine_data)

```

```python
passenger_vaccine_data = {
        'Name': [],
        'Vaccine': [],
        'Location': [] }
for _, passenger in titanic_data.iterrows():
     for _, row in vaccine_data.iterrows():
        if passenger['Age'] >= row['Age min'] and passenger['Age'] <= row['Age max']:
            passenger_vaccine_data['Name'].append(passenger['Name'])
            passenger_vaccine_data['Vaccine'].append(row['Vaccine'])
            passenger_vaccine_data['Location'].append(row['Location'])
pd.DataFrame(passenger_vaccine_data)

```
